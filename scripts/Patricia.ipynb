```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Simulate pre-storm VV backscatter (in dB)
# Water: -18 dB, Urban: -6 dB, Vegetation: -10 dB
pre_storm = np.random.normal(loc=[-18, -6, -10], scale=1.5, size=(100, 3))

# Simulate post-storm: water areas stay low, flooded urban/veg drop in dB
post_storm = pre_storm.copy()
post_storm[:, 1:] -= np.random.uniform(2, 6, size=(100, 2))  # Assume flooding lowers reflectance

# Combine into arrays
labels = ['Water', 'Urban', 'Vegetation']
pre_flat = pre_storm.flatten()
post_flat = post_storm.flatten()
classes = np.repeat(labels, 100)

# Plot pre vs post for each class
plt.figure(figsize=(10, 5))
for i, label in enumerate(labels):
    plt.hist(pre_storm[:, i] - post_storm[:, i], bins=20, alpha=0.6, label=label)

plt.title("Simulated VV Backscatter Drop Due to Flooding")
plt.xlabel("dB Drop (Pre - Post)")
plt.ylabel("Pixel Count")
plt.legend()
plt.grid(True)
plt.show()
```
