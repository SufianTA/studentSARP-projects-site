```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Simplified ODE model: 1 VOC → O₃
def reactions(y, t):
    voc, o3 = y
    k = 0.2  # VOC decay rate (hr⁻¹)
    r = 0.15  # O₃ formation rate from VOC (ppbv/hr)
    dvoc_dt = -k * voc
    do3_dt = r * voc
    return [dvoc_dt, do3_dt]

# Time: 0–24 hours post-fire
t = np.linspace(0, 24, 200)
y0 = [100, 0]  # 100 ppbv VOC, 0 ppbv O₃ initially

# Solve system
solution = odeint(reactions, y0, t)
voc, o3 = solution.T

# Plot results
plt.plot(t, voc, label="VOC", linewidth=2)
plt.plot(t, o3, label="O₃", linewidth=2)
plt.xlabel("Time (hours)")
plt.ylabel("Concentration (ppbv)")
plt.title("Simulated VOC Decay and Ozone Formation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
