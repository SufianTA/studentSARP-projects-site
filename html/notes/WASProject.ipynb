```python

# STEP 1: Import Required Libraries
!pip install pymannkendall
# We import all necessary Python libraries to load, analyze, and visualize the data.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm
import pymannkendall as mk  # For trend detection using Mann-Kendall Test
import warnings

# Suppress the datetime parsing warning
warnings.filterwarnings("ignore", message="Could not infer format*")

# Set a clean visual style for plots
sns.set(style="whitegrid")


# STEP 2: Load the CSV File

# Read the converted WAS input data from a CSV file into a DataFrame.
was_df = pd.read_csv("WASInput.csv")


# STEP 3: Clean Column Names

# Normalize and sanitize column names to make them easier to use in code.
was_df.columns = was_df.columns.str.strip()
was_df.columns = was_df.columns.str.replace(r'\s*\(.*?\)', '', regex=True)
was_df.columns = was_df.columns.str.replace('.', '', regex=False)
was_df.columns = was_df.columns.str.replace('#', 'number')
was_df.columns = was_df.columns.str.replace('/', '_')
was_df.columns = was_df.columns.str.replace(' ', '_')


# STEP 4: Preview and Explore the Data

# Basic overview of the dataset: shape, columns, data types, and statistics.
print("Shape of dataset:", was_df.shape)
print("\nColumns:\n", was_df.columns.tolist())
print("\nInfo:")
was_df.info()
print("\nSummary Stats:")
display(was_df.describe())


# STEP 5: Handle Missing and Duplicate Data

print("\nMissing Values:\n", was_df.isnull().sum()[was_df.isnull().sum() > 0])
print("\nDuplicate Rows:", was_df.duplicated().sum())

# Drop unnecessary or empty columns
was_df = was_df.loc[:, ~was_df.columns.str.contains('^Unnamed')]

# Remove duplicate rows
was_df = was_df.drop_duplicates()

# Drop columns with more than 95% missing values
threshold = 0.95
was_df = was_df.loc[:, was_df.isnull().mean() < threshold]

# Fill missing numeric values with the median
numeric_cols = was_df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    was_df[col] = was_df[col].fillna(was_df[col].median())

print("\nPost-cleaning shape:", was_df.shape)
print("Remaining missing values:", was_df.isnull().sum().sum())


# STEP 6: Histograms with Standard Normal Distribution Overlay

# This helps us compare our variable distribution with a "perfect bell curve"
selected_vars = ['CH4', 'CO', 'Ethanol']
for col in selected_vars:
    if col in was_df.columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(was_df[col], kde=False, stat='density', bins=30, label="Data", color='skyblue')
        mu, std = was_df[col].mean(), was_df[col].std()
        x = np.linspace(*plt.xlim(), 100)
        p = norm.pdf(x, mu, std)
        plt.plot(x, p, 'r', linewidth=2, label="Normal Dist")
        plt.title(f"Histogram and Normal Curve for {col}")
        plt.xlabel(col)
        plt.ylabel("Density")
        plt.legend()
        plt.show()
    else:
        print(f" Column '{col}' not found.")


# STEP 7: Correlation Heatmap and Pairplot

# Correlation heatmap reveals how gases relate to each other.
existing_cols = [col for col in selected_vars if col in was_df.columns]
if len(existing_cols) >= 2:
    plt.figure(figsize=(8, 6))
    sns.heatmap(was_df[existing_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.show()

    sns.pairplot(was_df[existing_cols])
    plt.suptitle("Pairplot of Selected Variables", y=1.02)
    plt.show()


# STEP 8: Save Cleaned Dataset

# Save cleaned data to file so we can use it later or share it.
was_df.to_csv("WAS_Cleaned_Output.csv", index=False)
print(" Cleaned WAS data saved to 'WAS_Cleaned_Output.csv'")


# STEP 9: Advanced Statistical Analysis

# Analyze each variable's shape using Skewness and Kurtosis.
# Skewness tells us about symmetry; Kurtosis about peak & tails.
stats_list = []
for col in selected_vars:
    if col in was_df.columns:
        col_skew = skew(was_df[col])
        col_kurtosis = kurtosis(was_df[col])
        stats_list.append({
            'Variable': col,
            'Skewness': col_skew,
            'Kurtosis': col_kurtosis
        })
stats_df = pd.DataFrame(stats_list)
print(" Skewness and Kurtosis for Selected Variables:")
display(stats_df)


# STEP 10: Advanced Visualizations


# Violin plots show distribution, median, and quartiles visually
for col in selected_vars:
    if col in was_df.columns:
        plt.figure(figsize=(8, 4))
        sns.violinplot(x=was_df[col], inner="quartile", color='lightblue')
        plt.title(f"Violin Plot of {col}")
        plt.show()

# Boxplot by location (if location exists)
if 'Location' in was_df.columns:
    for col in selected_vars:
        if col in was_df.columns:
            plt.figure(figsize=(12, 5))
            sns.boxplot(x='Location', y=col, data=was_df)
            plt.xticks(rotation=45, ha='right')
            plt.title(f"{col} by Location")
            plt.tight_layout()
            plt.show()

# Scatter plot of variable vs latitude
if 'latitude' in was_df.columns:
    for col in selected_vars:
        if col in was_df.columns:
            plt.figure(figsize=(8, 5))
            sns.regplot(x='latitude', y=col, data=was_df)
            plt.title(f"{col} vs Latitude")
            plt.xlabel("Latitude")
            plt.ylabel(col)
            plt.grid(True)
            plt.show()

# Heatmap of top correlated numeric columns
top_corr_cols = was_df.select_dtypes(include=np.number).corr().abs().sum().sort_values(ascending=False).head(20).index.tolist()
plt.figure(figsize=(12, 10))
sns.heatmap(was_df[top_corr_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Top Correlated Variables (Numeric)")
plt.show()

# Save descriptive stats
stats_summary = was_df.describe(include='all').transpose()
stats_summary.to_csv("WAS_Stats_Report.csv")
print(" Statistics report saved as 'WAS_Stats_Report.csv'")


# STEP 11: Mann-Kendall Trend Test (Temporal Trends)

# Explanation:
# This test checks if there is a statistically significant trend (upward/downward) over time for each variable.

if 'Time' in was_df.columns:
    print("\n Mann-Kendall Trend Test Results:")
    for col in selected_vars:
        if col in was_df.columns:
            try:
                result = mk.original_test(was_df[col])
                print(f"\n {col}:")
                print(f"â€¢ Trend: {result.trend}")
                print(f"â€¢ p-value: {result.p:.5f}")
                print(f"â€¢ Tau: {result.Tau:.3f}")
                print(f"â€¢ S: Not available in this test")  # Optional explanation
            except Exception as e:
                print(f" {col} â€” Error in trend test: {e}")



# STEP 12: Time Series Plots of Selected Variables

# Explanation:
# Now that we've ensured 'Time' is a datetime column, we can plot how each gas concentration changes over time.
# This helps us visually detect trends, spikes, or seasonal patterns.

# Ensure the Time column is parsed correctly
was_df['Time'] = pd.to_datetime(was_df['Time'], errors='coerce')

# Sort by time just in case
was_df = was_df.sort_values('Time')

# Plot each variable over time
for col in selected_vars:
    if col in was_df.columns:
        plt.figure(figsize=(12, 5))
        plt.plot(was_df['Time'], was_df[col], marker='o', linestyle='-', alpha=0.7, label=col)
        plt.title(f"{col} Over Time")
        plt.xlabel("Time")
        plt.ylabel(f"{col} Concentration")
        plt.grid(True)
        plt.tight_layout()
        plt.legend()
        plt.show()
    else:
        print(f" Column '{col}' not found for time series plot.")
```
