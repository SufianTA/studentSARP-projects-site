```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate fire spread data (Rate of Spread in m/min) for 3 wind conditions
np.random.seed(42)
low_wind = np.random.normal(loc=10, scale=3, size=(100, 100))
medium_wind = np.random.normal(loc=30, scale=6, size=(100, 100))
high_wind = np.random.normal(loc=60, scale=10, size=(100, 100))

# Optional: clip to realistic values and simulate burn zones
low_wind[low_wind < 0] = 0
medium_wind[medium_wind < 0] = 0
high_wind[high_wind < 0] = 0

scenarios = [low_wind, medium_wind, high_wind]
titles = ['Low Wind', 'Medium Wind', 'High Wind']

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
vmin, vmax = 0, 100

for i in range(3):
    im = axes[i].imshow(scenarios[i], cmap='hot', vmin=vmin, vmax=vmax)
    axes[i].set_title(titles[i])
    axes[i].axis('off')

# Add colorbar
cbar_ax = fig.add_axes([0.92, 0.25, 0.015, 0.5])
fig.colorbar(im, cax=cbar_ax, label='Rate of Spread (m/min)')

plt.suptitle("🔥 Simulated FlamMap Fire Spread under Varying Wind Speeds", fontsize=16)
plt.tight_layout(rect=[0, 0, 0.9, 0.95])
plt.show()
```
