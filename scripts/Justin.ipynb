```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated vertical profile data
altitudes = np.array([0, 250, 500, 750, 1000, 1250])

# NO3 profiles for 2025 (Summer) vs. 2021 (Winter)
no3_2025 = np.array([5, 8, 12, 10, 7, 3])
no3_2021 = np.array([6, 10, 14, 11, 9, 5])

# Boundary layer heights
bl_2025 = 750
bl_2021 = 1000

# Plot comparison
plt.figure(figsize=(6, 8))
plt.plot(no3_2025, altitudes, marker='o', label='NO Summer 2025', color='blue')
plt.plot(no3_2021, altitudes, marker='s', label='NO Winter 2021', color='green')
plt.axhline(bl_2025, color='blue', linestyle='--', label='BL Height 2025')
plt.axhline(bl_2021, color='green', linestyle='--', label='BL Height 2021')

plt.xlabel("NO Concentration (ppbv)")
plt.ylabel("Altitude (m)")
plt.title("NO Vertical Profiles with Boundary Layer Heights")
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()  # Invert to match flight altitude conventions
plt.tight_layout()
plt.show()
```
