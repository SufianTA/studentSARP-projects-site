```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate NDWI band data (values between -0.5 and 1)
np.random.seed(42)
ndwi = np.random.uniform(low=-0.5, high=1.0, size=(100, 100))

# Simulate oil slick mask (NDWI < 0.1 and some noise)
oil_mask = (ndwi < 0.1).astype(int)
noise = np.random.binomial(1, 0.05, size=ndwi.shape)
oil_mask = np.clip(oil_mask + noise, 0, 1)

# Visualization
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(ndwi, ax=axs[0], cmap="viridis", cbar_kws={"label": "NDWI"})
axs[0].set_title("Simulated NDWI (Sentinel-2)")
axs[0].axis('off')

sns.heatmap(oil_mask, ax=axs[1], cmap="Reds", cbar=False)
axs[1].set_title("Simulated Oil Slick Detection Mask")
axs[1].axis('off')

plt.suptitle("Synthetic NDWI and Oil Slick Classification (GEE-style)", fontsize=14)
plt.tight_layout()
plt.show()
```
